## About

This repository is a skeleton for a reproducible Python pipeline that allows you to create and publish analyses online hosted through Github Pages. It works by using `nbconvert` to export static HTML versions of `ipynb` files in the `notebooks` directory which can be hosted from the repository containing the original work. This is nice because it reduces the number of repositories related to a single project and allows you to keep all of your work in one place.

### How To Use It

To use this pipeline, there are a few simple steps.

1. Copy the contents of this repo locally. Do not edit any of the files in the `utils` or `export/menus` directories.
2. Put any analysis notebooks in the `notebooks` directory. Make sure your filenames follow the [naming scheme](#Naming-Scheme). Make sure each notebook includes the [JavaScript to generate the menus](#JavaScript-Menus).
3. Create a [homepage](#Project-Homepage) for your project.
4. Run the file `utils/export.py`. This will generate the necessary JavaScript files for the page menus and export each `ipynb` file in the `notebooks` directory. All output HTML files (except `index.html`) should appear in the `export` directory.
5. Push to Github and you're done!

### Naming Scheme

Files in the `notebooks` directory need to follow a particular naming scheme so that the Python scripts which export the notebooks order them correctly in the menu. This scheme begins each file with a 2-digit number indicating the order (first is `01`, second `02`, etc.). Each word in the notebook title (what you want displayed in the menu) should follow that number with hyphens in place of spaces. 

For example, if the fifth notebook was titled "Principal Component Analysis With Scree Plot", then the filename would be `05-principal-component-analysis-with-scree-plot.ipynb`. The utilities function will automatically capitalize all words in the title.

#### Project Homepage

The homepage for your project will also be located in this folder but it will be named `index.ipynb` (not following the naming scheme). This is important because Github pages requires an `index.html` file in the root directory for the project as the homepage.

A sample structure, from the [Github repo for this project](https://github.com/chrispyles/python-to-github/tree/master/notebooks), is given below.

```
| notebooks
  | 01-sample-page.ipynb
  | index.ipynb
```

### JavaScript Menus

The menus in this project are generated through Python writing a JavaScript file which is executed in the HTML output files. In order to include the menu on each page, the Rmd file needs to include an HTML `script` tag so that the menu is loaded with in the knitted document.

For all notebooks _except the index notebook_, include the HTML below at the top of the notebook (I suggest below the page title and author, as above):

```html
<script type="text/javascript" src="menus/menu.js"></script>
```

The `utils/generate-menu.py` script, executed in the `utils/export.py` script, creates the JS file referenced above, along with the menu for the index page.

For the index notebook, the HTML `script` tag is a bit different:

```html
<script type="text/javascript" src="export/menus/index-menu.js"></script>
```

This is because the `index.html` file is in a different directory from the other HTML files and so the links to other pages are different.
