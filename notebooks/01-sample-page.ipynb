## Dependencies and Data

To begin, let's load the dependencies and the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris = pd.read_csv("../data/iris.csv")
iris.head()
```

## Principal Component Analysis

Let's do a quick PCA of the four numerical data sets.

```python
X = iris.iloc[:,0:4].values
X = (X - np.mean(X, axis=0)) / np.sqrt(iris.shape[0])

U, Sigma, Vt = np.linalg.svd(X, full_matrices=False)
```

Before we draw any conclusions, let's do a scree plot of the singular values.

```python
sns.lineplot(range(1, 5), Sigma**2)
plt.title("Scree Plot of Iris Data")
plt.xlabel("Principal Component Number")
plt.xticks(range(1, 5))
plt.ylabel(r"$\Sigma_i^2$");
```

Now let's look at the scatter plot of the first two PCs, differentiating hue by the type of iris

```python
first_2 = X @ Vt.T[:,0:2]
sns.scatterplot(first_2[:,0], first_2[:,1], hue = iris["species"])
plt.title("PC 2 vs. PC 1 for Iris Species")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2");
```

As we can see from the plot above, PC1 separates the setosas from the versicolors and virginicas.
